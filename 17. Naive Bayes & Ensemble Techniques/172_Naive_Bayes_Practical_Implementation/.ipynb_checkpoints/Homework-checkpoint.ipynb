```python
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import BernoulliNB, MultinomialNB, GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.naive_bayes import BernoulliNB , MultinomialNB , GaussianNB
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# 1. Load the Spambase dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/spambase/spambase.data"
column_names = [
    "word_freq_make", "word_freq_address", "word_freq_all", "word_freq_3d", "word_freq_our",
    "word_freq_over", "word_freq_remove", "word_freq_internet", "word_freq_order",
    "word_freq_mail", "word_freq_receive", "word_freq_will", "word_freq_people",
    "word_freq_report", "word_freq_addresses", "word_freq_free", "word_freq_business",
    "word_freq_email", "word_freq_you", "word_freq_credit", "word_freq_your", "word_freq_font",
    "word_freq_000", "word_freq_money", "word_freq_hp", "word_freq_hpl", "word_freq_george",
    "word_freq_650", "word_freq_lab", "word_freq_labs", "word_freq_telnet", "word_freq_857",
    "word_freq_data", "word_freq_415", "word_freq_85", "word_freq_technology",
    "word_freq_1999", "word_freq_parts", "word_freq_pm", "word_freq_direct", "word_freq_cs",
    "word_freq_meeting", "word_freq_original", "word_freq_project", "word_freq_re",
    "word_freq_edu", "word_freq_table", "word_freq_conference", "char_freq_;", "char_freq_(",
    "char_freq_[", "char_freq_!", "char_freq_$", "char_freq_#", "capital_run_length_average",
    "capital_run_length_longest", "capital_run_length_total", "is_spam"
]

df = pd.read_csv(url, header=None, names=column_names)
```

```python
# 2. Separate features and target
X = df.drop("is_spam", axis=1)
y = df["is_spam"]
```

## Applying different Naive Bayesian Techniques 
 1. Bernoulli Naive Bayes 
 2. Multinomial Naive Bayes 
 3. Gaussian Naive Bayes

### 1. Bernoulli Naive Bayes

```python
bernoulli = BernoulliNB()

accuracy_bernoulli = cross_val_score(bernoulli , X , y , cv = 10 , scoring = 'accuracy').mean()
precision_bernoulli = cross_val_score(bernoulli , X , y , cv = 10 , scoring = 'precision').mean()
recall_bernoulli = cross_val_score(bernoulli , X , y , cv = 10 , scoring = 'recall').mean()
f1_bernoulli = cross_val_score(bernoulli , X , y , cv = 10 , scoring='f1').mean()
```

```python
print("Bernoulli Naive Bayes Classifier:")
print(f"Accuracy: {accuracy_bernoulli}")
print(f"Precision: {precision_bernoulli}")
print(f"Recall: {recall_bernoulli}")
print(f"F1 Score: {f1_bernoulli}")
```

### 2. Multinomial Naive Bayes

```python
multinomial = MultinomialNB()

accuracy_multinomial = cross_val_score(multinomial , X , y , cv = 10 , scoring = 'accuracy').mean()
precision_multinomial = cross_val_score(multinomial , X , y , cv = 10 , scoring = 'precision').mean()
recall_multinomial = cross_val_score(multinomial , X , y , cv = 10 , scoring = 'recall').mean()
f1_multinomial = cross_val_score(multinomial , X , y , cv = 10 , scoring='f1').mean()
```

```python
print("Multinomial Naive Bayes Classifier:")
print(f"Accuracy: {accuracy_multinomial}")
print(f"Precision: {precision_multinomial}")
print(f"Recall: {recall_multinomial}")
print(f"F1 Score: {f1_multinomial}")
```

### 3. Gaussian Naive Bayes

```python
gaussian = GaussianNB()

accuracy_gaussian = cross_val_score(gaussian , X , y , cv = 10 , scoring = 'accuracy').mean()
precision_gaussian = cross_val_score(gaussian , X , y , cv = 10 , scoring = 'precision').mean()
recall_gaussian = cross_val_score(gaussian , X , y , cv = 10 , scoring = 'recall').mean()
f1_gaussian = cross_val_score(gaussian , X , y , cv = 10 , scoring='f1').mean()
```

```python
print("Gaussian Naive Bayes Classifier:")
print(f"Accuracy: {accuracy_gaussian}")
print(f"Precision: {precision_gaussian}")
print(f"Recall: {recall_gaussian}")
print(f"F1 Score: {f1_gaussian}")
```

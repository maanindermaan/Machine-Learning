```python
data1 = pd.DataFrame({
    'key1' : [1,2,4,5,6],
    'key2' : [4,5,6,7,8],
    'key3' : [3,4,5,6,6]
})
```

```python
data1
```

```python
data2 = pd.DataFrame({
    'key1' : [1,2,45,6,67],
    'key4' : [56,5,6,7,8],
    'key5' : [3,56,5,6,6]
})
```

```python
data2
```

```python
data1
```

```python
data2
```

```python
# MERGE OPERATION()
```

```python
pd.merge(data1 , data2)
```

```python
# How inner merge is working ?
# ye dono df1 and df2 ko dekh rha hai. Ab dono mein column konsa common hai, key1 is common in both ab iske andar.
# Ab key1 ke andar '1' jha pe avaiblabe hai key2 ke andar '1' bhi same position pe available hai, ab 4 , 3 , 56 and 3 ko lo or yha dedo
# and same for every other

# jo bhi aapka common column condition hai, isko yha pe le lerha hai or uske baad jo bhi data available hai in common usko nikal lo.
# ab jaise 4 d1 mein hai but d2 mein nahi hai toh kuch bhi nahi leke aaya and same goes for 5
```

```python
pd.merge(data1 , data2 , how = 'left')
```

```python
# jo bhi data aapne function mein pehle likha hai use left i.e. pehle lelega
# left means ki sab kuch uthao left se or tab jo bhi common hai right wale mein (data2) mein, agar nahi hai available toh NaN mardo
# sab kuch uthao left se or match kro right se 
```

```python
# sab kuch uthao right se or match kro left se 
pd.merge(data1 , data2 , how = 'right')
```

```python
pd.merge(data1 , data2 , how = 'outer')
# ab outer kroge toh yeah left se bhi saara utha ke layega or right se bhi saara utha ke layega
# jha match hoga vha data dega jha match ni hoga vha left hand side , right hand size dono jagah NaN dedega
```

```python
pd.merge(data1 , data2 , how = 'cross')
# matlab sabko sabse khojo or sabko sabse multiply karo nikalke lao(cartesian product)
```

```python
# jisse match karwana chahte ho vo hijega
pd.merge(data1 , data2 , how = 'outer' , on = 'key1')
```

```python
# JOIN OPERATION
```

```python
data1 = pd.DataFrame({'key1':[1,2,4,5,6],
                      'key2':[4,5,6,7,8],
                      'key3':[3,4,5,6,6]},
                     index = ['a','b','c','d','e']
)
```

```python
data2 = pd.DataFrame({'key11':[1,2,4,5,6],
                      'key22':[4,5,6,7,8],
                      'key33':[3,4,5,6,6]
},index=['a','b','h','i','j']
)
```

```python
data1.join(data2)
# this is similar to merge 
# in merge -> vo column ko check krrha hai or column ke data ko check krke agar same hai toh map karwa rha hai
# in join -> isme vo index pe kaam karta hai or agar same hai tabhi vo vo join krta hai otherwise NaN,
```

```python
# By default join is left join
```

```python
data1.join(data2 , how = 'right')
```

```python
# dono mein jo common hai vo dikhadega aapko
data1.join(data2 , how = 'inner')
```

```python
# dono ka intersection dikha dega.
# left bhi pura dikhayenge , right bhi pura dikhayenge jha pe match krega data nahi match kia toh NaN
data1.join(data2 , how = 'outer')
```

```python
# cross mein ek ek ka mapping
data1.join(data2 , how = 'cross')
```

```python
# dono same hi hai bas yehi difference hai ki merge works on column values and join works on index values
```

```python
# jha bhi fare 10 euro se niche hai usko likhdo cheapest one 10-20 hai toh mid and 20>uper hai toh high
```

```python
df['Fare_status'] = df['Fare'].apply(lambda x: "low" if x < 10 else ("mid" if 10 <= x <= 20 else "high"))
```

```python
def cat_fare(x):
    if x<10 :
        return "cheap"
    elif x>=10 and x<20:
        return "mid"
    else:
        return "high"
```

```python
df['cat_fare'] = df['Fare'].apply(cat_fare)
```

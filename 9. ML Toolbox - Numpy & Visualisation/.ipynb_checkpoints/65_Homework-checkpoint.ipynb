```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

Q1. Using this generated data plot a scatter plot.

```python
plt.scatter(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Scattered Data')
plt.show()
```

Q2. Draw four line plots using the subplot() function Use the following data:

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for the four lines
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots and plot the data on them
plt.figure(figsize=(10, 8))

# Plot for Line 1
plt.subplot(2, 2, 1)
plt.plot(x1, y1, label='Line 1', color='blue')
plt.title('Line 1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Plot for Line 2
plt.subplot(2, 2, 2)
plt.plot(x2, y2, label='Line 2', color='green')
plt.title('Line 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Plot for Line 3
plt.subplot(2, 2, 3)
plt.plot(x3, y3, label='Line 3', color='red')
plt.title('Line 3')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Plot for Line 4
plt.subplot(2, 2, 4)
plt.plot(x4, y4, label='Line 4', color='purple')
plt.title('Line 4')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Adjust layout for better presentation
plt.tight_layout()

# Show the plots
plt.show()
```

Q3. Using the following data plot a bar plot and a horizontal bar plot.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

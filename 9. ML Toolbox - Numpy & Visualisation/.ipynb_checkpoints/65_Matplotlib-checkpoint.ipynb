```python
# controlling scatter
colours = np.random.rand(50)
sizes = 1000*np.random.rand(50)
plt.scatter(x,y , c = colours , s = sizes)
```

```python
# intensity
colours = np.random.rand(50)
sizes = 1000*np.random.rand(50)
plt.scatter(x,y , c = colours , s = sizes , alpha = 0.5) 
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("random data")
```

```python
# You can take any value of colour using colour coding
x = np.random.rand(50)
y = np.random.rand(50)
colours = np.random.rand(50)
sizes = 1000*np.random.rand(50)
plt.scatter(x,y , c = '#A64D13' , s = sizes , alpha = 0.5)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("random data")
```

```python
# Marker
x = np.random.rand(50)
y = np.random.rand(50)
colours = np.random.rand(50)
sizes = 1000*np.random.rand(50)
plt.scatter(x,y , c = '#A64D13' , s = sizes , alpha = 0.5 , marker = 'x')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("random data")
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z)
plt.show()
```

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
from bokeh.plotting import figure ,output_file , show
from bokeh.sampledata.iris import flowers

output_file('68test.html') # jha ouput dega yeah

p = figure(title = "test flower")
p.xaxis.axis_label = "patel length"
p.yaxis.axis_label = "patel width"
p.circle(flowers['petal_length'] , flowers['petal_width'])
show(p)
```

```python
from bokeh.plotting import figure ,output_file , show
from bokeh.sampledata.iris import flowers

output_file('68line.html')

p = figure(title = "line graph")
p.xaxis.axis_label = "x axis"
p.yaxis.axis_label = "y axis"
p.line(flowers['petal_length'] , flowers['petal_width'])
show(p)
```

```python
x = [2,3,4,5,6]
y = [3,4,5,6,7]
output_file("68_line_2.html")
p = figure(title = "line plot")
p.line(x,y)
show(p)
```

```python
x = [2,3,4,5,6]
y = [3,4,5,6,7]
output_file("68_line_3.html")
p = figure(title = "line plot")
p.scatter(x,y,fill_color = 'red' , legend_label = 'red points',size = 20)
show(p)
```

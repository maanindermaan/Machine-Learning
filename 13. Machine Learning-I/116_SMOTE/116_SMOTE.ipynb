## SMOTE(Synthetic Minority Oversampling Technique)
SMOTE (Synthetic Minority Over-sampling Technique) is a technique used in machine learning to address imbalanced datasets where the minority class has significantly fewer instances than the majority class. SMOTE involves generating synthetic instances of the minority class by interpolating between existing instances.

2 minority ke bech mein line daldenge and jitne bhi points aa sakte hai unke bech mein hum vo lagadenge(Interpolation -> generating data points). Upsampling mein simply hum ek ke uper ek data points dalte the

```python
# Ye library imbalanced dataset banane mein help krega and it is very useful as well
from sklearn.datasets import make_classification
```

```python
# X independent feature
# Y dependent feature

X,Y = make_classification(n_samples = 1000, n_features = 2 ,n_redundant=0, n_clusters_per_class= 1, weights = [0.90],random_state=1)
```

```python
# converting into data frames
import pandas as pd
```

```python
df1 = pd.DataFrame(X , columns = ['f1' , 'f2'])
df2 = pd.DataFrame(Y , columns = ['target'])
final_df = pd.concat([df1 , df2] , axis = 1)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
# transform the data set
# ab SMOTE import krne ke baad mai apne data ko tranform kardunga kyunki ab merko jo minority data points tha merko vo majority ke hisab se lana hai through SMOTE
```

```python
# initialising smote
oversample = SMOTE()
# ab fit_resample ke andar i have to give data points and target feature
oversample.fit_resample(final_df[['f1' , 'f2']] , final_df['target'])
```

```python
X , Y = oversample.fit_resample(final_df[['f1' , 'f2']] , final_df['target'])
# X-> Jo mere same feautres hai uske sath ek numpy array dega (independent features)
# Y-> Jo aapka quantities hai numbers hai 

# ab humne dekha tha uper 0 894 tha and 1 106 tha ab after this SMOTING technique 0,1 dono 894 hojayenge
```

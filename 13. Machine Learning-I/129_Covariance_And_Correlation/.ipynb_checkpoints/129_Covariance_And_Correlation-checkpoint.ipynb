Q3. Suppose you are conducting a study to examine the relationship between the number of hours of
exercise per week and body mass index (BMI) in a sample of adults. You collected data on both variables
for 50 participants. Calculate the Pearson correlation coefficient and the Spearman's rank correlation
between these two variables and compare the results.

```python
data = {
    "Exercise Hours": [3, 5, 2, 4, 6, 3, 2, 4, 5, 1, 3, 6, 4, 5, 2, 1, 4, 5, 3, 2, 6, 1, 3, 4, 2, 5, 3, 4, 6, 2, 5, 3, 4, 1, 6, 3, 2, 4, 5, 3, 1, 2, 5, 4, 6, 3, 2, 1],
    "BMI": [25.0, 26.5, 27.2, 29.1, 24.8, 28.5, 30.3, 26.7, 23.8, 31.2, 29.8, 24.9, 27.6, 30.0, 25.5, 30.7, 29.3, 25.9, 28.2, 26.0, 30.1, 27.4, 29.9, 27.8, 31.5, 25.7, 30.9, 28.8, 26.4, 29.7, 25.2, 27.1, 28.6, 31.8, 25.6, 29.2, 30.4, 26.8, 24.7, 29.0, 27.3, 25.4, 31.3, 26.1, 30.8, 28.7, 24.6, 27.9]
}

df3 = pd.DataFrame(data)

# Calculate Pearson correlation coefficient
pearson_corr = df3["Exercise Hours"].corr(df3["BMI"], method='pearson')

# Calculate Spearman's rank correlation coefficient
spearman_corr = df3["Exercise Hours"].corr(df3["BMI"], method='spearman')

df3.corr(method = 'pearson')
df3.corr(method = 'spearman')
```

```python
df3
```

```python
pearson_corr
```

```python
spearman_corr
```

Q4. A researcher is interested in examining the relationship between the number of hours individuals
spend watching television per day and their level of physical activity. The researcher collected data on
both variables from a sample of 50 participants. Calculate the Pearson correlation coefficient between
these two variables.

```python
data = {
    "TV Hours": [2, 5, 6, 2, 4, 5, 3, 2, 4, 6, 5, 1, 3, 2, 4, 5, 3, 2, 4, 6, 5, 1, 3, 4, 2, 5, 6, 2, 4, 5, 3, 2, 4, 6, 5, 1, 3, 2, 4, 5, 3, 2, 4, 6, 5, 1],
    "Physical Activity": [5, 4, 6, 3, 2, 6, 4, 3, 5, 6, 4, 2, 3, 1, 5, 6, 4, 3, 5, 6, 4, 2, 3, 1, 5, 6, 4, 3, 5, 6, 4, 2, 3, 1, 5, 6, 4, 3, 5, 6, 4, 2, 3, 1, 5, 6]
}

df4 = pd.DataFrame(data)
```

```python
df4
```

```python
df4.corr(method = 'pearson')
```

```python
# Calculate the correlation matrix with Pearson method
correlation_matrix = df4.corr(method='pearson')

# Extract the Pearson correlation coefficient between TV Hours and Physical Activity
tv_hours_activity_corr = correlation_matrix.loc["TV Hours", "Physical Activity"]

# Print the correlation coefficient
print(f"Pearson Correlation Coefficient: {tv_hours_activity_corr:.2f}")
```

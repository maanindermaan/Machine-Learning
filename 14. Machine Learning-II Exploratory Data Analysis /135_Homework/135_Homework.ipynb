Q2. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of
features. What is the minimum number of principal components required to explain 90% of the variance in
the data?

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Separate features and target variable(This is not given so we take any one randomly)
```

```python
df1.head(1)
```

```python
# suppose qaulity is target variable
```

```python
# 1. Separate features and target variable
X = df1.drop('quality' , axis=1)
y = df1['quality']
```

```python
# 2. Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# 3. Apply PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)
```

```python
# 4. Calculate cumulative explained variance ratio
cummulative_variance = np.cumsum(pca.explained_variance_ratio_)
```

```python
# 5. Find the number of components explaining 90% variance
n_components_90 = np.argmax(cummulative_variance >= 0.90) + 1 
# Adding 1 because indexing starts from 0
```

##### np.argmax( ): 
This function returns the indices of the maximum values along a given axis. In the context of PCA, it's used to find the index (number of components) at which the cumulative explained variance first exceeds or equals a certain threshold (in this case, 0.90 or 90% of the variance). It helps to identify the minimum number of principal components needed to explain a specified percentage of variance in the dataset.
##### np.cumsum( ):
This function computes the cumulative sum of array elements along a specified axis. In PCA, pca.explained_variance_ratio_ contains the variance explained by each of the principal components. np.cumsum() is used to calculate the cumulative explained variance ratio. It helps to understand how much variance is explained by each additional principal component.

```python
# create a copy
df_copy = df.copy()
```

```python
df_copy
```

```python
df_copy = df_copy.drop(df_copy.index[10472])
```

```python
df_copy.shape
```

```python
df_copy['Reviews'] = df_copy['Reviews'].astype(int)
```

```python
df_copy.info()
```

```python
df_copy['Size'].unique()
```

```python
# converting mb -> kb
```

```python
df_copy['Size'] = df_copy['Size'].str.replace('M','000')
df_copy['Size'] = df_copy['Size'].str.replace('k','')
df_copy['Size'] = df_copy['Size'].replace('Varies with device',np.nan)
# Now you have to figure out how will you see these values as it have one value mainly called varies with device you have to keep a check on your own 
df_copy['Size'] = df_copy['Size'].astype(float)
```

```python
df_copy.info()
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
chars_to_remove = ['+' , ',' , '$']
cols_to_clean = ['Installs' , 'Price']
for item in chars_to_remove:
    for cols in cols_to_clean:
        df_copy[cols] = df_copy[cols].str.replace(item , '')
```

```python
df_copy['Installs']=df_copy['Installs'].astype('int')
df_copy['Price']=df_copy['Price'].astype('float')
```

```python
df_copy.info()
```

```python
df_copy['Last Updated']
```

```python
df_copy['Last Updated'] = pd.to_datetime(df_copy['Last Updated'])
df_copy['Day'] = df_copy['Last Updated'].dt.day
df_copy['Month'] = df_copy['Last Updated'].dt.month
df_copy['Year'] = df_copy['Last Updated'].dt.year
```

```python
df_copy.head()
```

```python
df_copy.duplicated("App")
```

```python
df_copy[df_copy.duplicated("App")]
```

```python
df_copy = df_copy.drop_duplicates(subset=['App'] , keep='first')
```

```python
# No duplicates now
df_copy[df_copy.duplicated("App")].shape
```

```python
numeric_feature = [feature for feature in df_copy.columns if df_copy[feature].dtype != 'O']
categorical_feature = [feature for feature in df_copy.columns if df_copy[feature].dtype == 'O']
```

### KDEPlot -> Numeric Features

```python
plt.figure(figsize=(15,15))
plt.suptitle('Univariate Analysis of Numeric Features' , fontsize=20 , fontweight='bold', alpha=0.8, y=1.)

for i in range(0 , len(numeric_feature)):
    plt.subplot(5,3,i+1)
    sns.kdeplot(x=df_copy[numeric_feature[i]] , shade = True , color='r')
    plt.xlabel(numeric_feature[i])
```

### CATPlot -> Categorical Features 

```python
plt.figure(figsize=(20, 15))
plt.suptitle('Univariate Analysis of Categorical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)
category = [ 'Type', 'Content Rating']
for i in range (0 , len(category)):
    plt.subplot(2,2,i+1)
    sns.countplot(x=df[category[i]] , palette='Set2')
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
    
# for i in range(0, len(category)):
#     plt.subplot(2, 2, i+1)
#     sns.countplot(x=df[category[i]],palette="Set2")
#     plt.xlabel(category[i])
#     plt.xticks(rotation=45)
#     plt.tight_layout()
```

```python
# 1. Date of Journey
df['Date_of_Journey'].str.split('/')
```

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head(2)
```

```python
# Converting to numericals
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.drop('Date_of_Journey' , axis=1 , inplace=True)
```

```python
df.head(2)
```

```python
df.info()
```

```python
# 2. Arrival Time Now
df['Arrival_Time'].str.split(' ').str[0]
```

```python
df['Arrival_Time'].str.split(' ').str[0].str.split(':')
```

```python
df['Arrival_Hours'] = df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[0]
df['Arrival_Minutes'] = df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[1]
df.drop('Arrival_Time' , axis=1 , inplace=True)
```

```python
df.head(2)
```

```python
df.info()
```

```python
df['Arrival_Hours'] = df['Arrival_Hours'].astype(int)
df['Arrival_Minutes'] = df['Arrival_Minutes'].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
# 3. Departure Time
```

```python
df['Dep_Hours'] = df['Dep_Time'].str.split(':').str[0].astype(int)
df['Dep_Minutes'] = df['Dep_Time'].str.split(':').str[1].astype(int)
df.drop('Dep_Time' , inplace=True , axis=1)
```

```python
df.head(2)
```

```python
df.info()
```

```python
df.drop('Route' , axis=1, inplace=True)
```

```python
df.head(2)
```

```python
# 4. Duration
```

```python
df['Duration_Hours'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration_Minutes'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df.drop('Duration' , axis=1, inplace=True)
```

What is the average flight duration in the Flight Price dataset?

```python
df.describe()
```

```python
df.head()
```

```python
df['Total_Duration_Minutes'] = df['Duration_Hours'] * 60 + df['Duration_Minutes']
```

```python
df.head()
```

```python
average_duration_minutes = df['Total_Duration_Minutes'].mean()
```

```python
average_duration_minutes
```

```python
hours = int(average_duration_minutes / 60)
minutes = int(average_duration_minutes % 60)

print(f"The average flight duration is approximately {hours} hours and {minutes} minutes.")
```

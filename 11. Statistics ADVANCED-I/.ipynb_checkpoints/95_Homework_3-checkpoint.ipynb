Q2. Write a Python script to conduct a hypothesis test on the difference between two population means,
given a sample from each population.

```python
import scipy.stats as stats

sample1 = [25, 30, 35, 40, 45]  
sample2 = [20, 28, 32, 38, 42] 

t_val , p_val = stats.ttest_ind(sample1 , sample2)

alpha = 0.05

if p_val < alpha:
    print("We reject the NULL hypthesis")
else:
    print("We failed to reject the NULL hypthesis")
```

Q3. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom
parameter set to 10.

```python
import matplotlib.pyplot as plt
import numpy as np

dof = 10

sample = np.linspace(-4 , 4 , 400)

pdf = stats.t.pdf(sample , df = dof)

plt.figure(figsize=(8,6))
plt.plot(sample , pdf , color = 'blue')
plt.xlabel('Sample')
plt.ylabel('Probability Density')
plt.title('T-DISTRIBUTION PLOT')

plt.grid(visible=True)
plt.show()
```

Q4. Write a Python program to calculate the two-sample t-test for independent samples, given two
random samples of equal size and a null hypothesis that the population means are equal.

```python
import numpy as np
import scipy.stats as stats

# Generate two random samples of equal size
sample_size = 30
sample1 = np.random.normal(loc=50, scale=10, size=sample_size)  # Replace with your data
sample2 = np.random.normal(loc=55, scale=10, size=sample_size)  # Replace with your data

# Perform a two-sample t-test (assuming equal variances)
t_stat, p_value = stats.ttest_ind(sample1, sample2)

# Define the significance level (alpha)
alpha = 0.05

# Determine whether to reject the null hypothesis
if p_value < alpha:
    print("Reject the null hypothesis: The population means are not equal.")
else:
    print("Fail to reject the null hypothesis: The population means are equal.")

# Display the t-statistic and p-value
print(f"t-statistic: {t_stat}")
print(f"P-value: {p_value}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
dataset = load_iris()
```

```python
print(dataset.DESCR)
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
X = df.iloc[:,:-1]
y = dataset.target
```

```python
X,y
```

```python
# train test split
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.33 , random_state=42)
```

```python
# decision tree
from sklearn.tree import DecisionTreeClassifier
```

```python
# post pruning
treeclassifier = DecisionTreeClassifier()
```

```python
treeclassifier.fit(X_train , y_train)
```

```python
X_train.head()
```

```python
from sklearn import tree
plt.figure(figsize = (15,20))
tree.plot_tree(treeclassifier , filled = True)
```

```python
treeclassifier = DecisionTreeClassifier(criterion='entropy' , max_depth=2)
```

```python
treeclassifier.fit(X_train , y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(7,7))
tree.plot_tree(treeclassifier , filled = True)
```

```python
# predictions
y_pred = treeclassifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score , classification_report 
```

```python
score = accuracy_score(y_pred , y_test)
score
```

```python
print(classification_report(y_pred , y_test))
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
dataset = load_breast_cancer()
```

```python
dataset.feature_names
```

```python
print(dataset.DESCR)
```

```python
dataset.target
```

```python
pd.DataFrame(dataset.data , columns = dataset.feature_names)
```

```python
X = pd.DataFrame(dataset.data)
y = dataset.target
```

```python
X , y
```

```python
y.shape
```

```python
X.shape
```

```python
# 1. Train test split
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.33 , random_state=42)
```

```python
# 2. Making model
from sklearn.tree import DecisionTreeClassifier
treeclassifier = DecisionTreeClassifier()
treeclassifier.fit(X_train , y_train)
```

```python
# 3. Making tree
from sklearn import tree
plt.figure(figsize = (25,25))
tree.plot_tree(treeclassifier , filled = True)
```

```python
# 4. Making tree more shorter using Post Pruning Method
from sklearn.tree import DecisionTreeClassifier
treeclassifier = DecisionTreeClassifier(criterion='gini' , max_depth=2)
treeclassifier.fit(X_train , y_train)

from sklearn import tree
plt.figure(figsize = (10,10))
tree.plot_tree(treeclassifier , filled = True)
```

```python
# 5. Predictions
y_pred = treeclassifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score , classification_report
```

```python
accuracy_score(y_pred , y_test)
```

```python
print(classification_report(y_pred , y_test))
```
